```python
df['Embarked'] = df['Embarked'].replace({
    'C': 'Cherbourg',
    'Q': 'Queenstown',
    'S': 'Southampton'
})
```

```python
print(df['Embarked'])
```

```python
df.head()
```

```python
encoded_columns = pd.get_dummies(df['Embarked'], prefix='Embarked')
df = pd.concat([df, encoded_columns], axis=1)
df = df.drop(columns=['Embarked'])
```

```python
df.head()
```

```python
mean_survival_rates = df.groupby('Sex')['Survived'].mean()
print("Mean Survival Rates by Gender:")
print(mean_survival_rates)
```

```python

# Bar plot for mean survival rates
plt.subplot(1, 2, 1)
sns.barplot(x=mean_survival_rates.index, y=mean_survival_rates.values)
plt.title('Mean Survival Rates by Gender')
plt.ylabel('Survival Rate')
plt.xlabel('Gender')

# Box plot for survival distribution by gender
plt.subplot(1, 2, 2)
sns.boxplot(x='Sex', y='Survived', data=df)
plt.title('Survival Distribution by Gender')
plt.ylabel('Survival')
plt.xlabel('Gender')

plt.tight_layout()
plt.show()
```

```python
null_df = df.isnull().sum()
print(null_df)
```

```python
g = sns.catplot(
    data=df,
    x='Sex',
    y='Survived',
    hue='Sex',
    kind='bar',
    palette='pastel',
    height=5,
    aspect=1
)

# Customize the titles and axes
g.set_titles('{col_name}')
g.set_axis_labels('Gender', 'Mean Survival Rate')
g.set_xticklabels(['Female', 'Male'])
g.fig.suptitle('Survival Rates by Gender and Port of Embarkation', y=1.02)

# Show plot
plt.show()
```
